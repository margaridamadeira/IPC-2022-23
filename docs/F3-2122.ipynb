Queremos analisar algumas sequências de DNA e determinar o rácio transição/transcrição de pares dessas sequências ([Fonte](https://rosalind.info/problems/tran/)).

Essas sequências estão no formato FASTA. Uma sequência no formato FASTA começa com uma linha de descrição seguida por linhas com sequência de dados. A linha de descrição distingue-se das com sequências de dados por iniciar com o carácter `>` (maior-que).

As sequências de dados que vamos analisar têm todas a mesma dimensão e só contêm as bases `A`, `G` (que são purinas) e `C` e `T` (que são pirimidinas). Uma transição corresponde à substituição duma purina ou pirimidina por outra purina ou pirimidina, respetivamente; uma transversão corresponde à substituição de uma purina por uma pirimidina ou vice-versa.

Uma das formas de avaliar mutações pontuais entre duas sequências de DNA recorre ao cálculo do rácio transição/transcrição, que se obtém dividindo o número total de transições pelo número total de transcrições.

Queremos ler da consola um determinado número de sequências no formato FASTA que constituirá a nossa base de trabalho. Para simplificar, as nossas sequências FASTA têm apenas uma linha de dados.

Depois leremos também da consola, um número indeterminado de linhas com a identificação de duas sequências. Para cada linha lida, apresentaremos os identificadores das sequências e o rácio transição/transversão.



### Tarefa B

O input contém na primeira linha um número inteiro $n$ que representa o número de sequências FASTA a considerar. Seguem-se $n$ pares de linhas, em que cada par de linhas contém a descrição da sequência e a sequência. Depois existirão um número indeterminado de linhas cada uma contendo dois identificadores FASTA separados pelo carácter ' ' (espaço).

Para cada par de identificadores das sequências: caso o rácio não possa ser calculado, não deve ser apresentado qualquer output; caso seja possível o cálculo do rácio transcrição/transversão, devem ser apresentados os identificadores das sequências em causa e o valor do rácio apresentado como um número real com três casas decimais, sendo cada elemento separado dos demais pelo carácter ' ' (espaço).

## Processo

Referência, amostra do sensor e diferença
```

Queremos controlar um processo, com um sensor e um atuador. Sabemos qual o valor de referência pretendido mas ainda não temos um modelo. Além disso, a resolução do sensor e do atuador são diferentes. 

A forma como pretendemos implementar a abordagem inicial é a seguinte: começamos por ler o valor de referência e a resolução do atuador; depois vamos lendo os valores do sensor e decidimos de forma a compensar a distância ao valor de referência; se for adequado, podemos comandar o atuador usando múltiplos da resolução.

Por exemplo, queremos manter o valor (ie, a referência) a 20 e um atuador com resolução de 3. 
Ao lermos 30, comandamos o atuador com -9; na leitura seguinte obtemos 24, atuamos com -3; na leitura seguinte obtemos 16, atuamos com 3; obtemos em seguida 18 e indicamos 0, decidindo não atuar, porque a diferença é menor ou igual à resolução.


### Tarefa C

Faça um programa em Python para o controlo do processo descrito acima. Pretende-se pois determinar para cada leitura do valor corrente, qual o maior múltiplo do valor do atuador cabe na diferença entre a referência e o valor corrente; se a resolução for menor ou igual à diferença entre a referência e o valor corrente optamos por não alterar, ou seja, atuador a zero.

O seu programa começará por ler da consola dois números inteiros separados por espaço que correspondem à referência e resolução, respetivamente. Depois, o seu programa deverá ler da consola e até ao fim do input, um número indeterminado de linhas, cada uma contendo um número inteiro que corresponde ao valor corrente do sensor. Para cada linha lida com o valor corrente do sensor, o seu programa escreverá numa linha o valor inteiro do atuador.


**Casos de teste**

**Input 1**

```
